```python
def LoG(x,y, std):
    return ((x**2/std**4 - 1/std**2)+(y**2/std**4 - 1/std**2)) *np.exp(-(x**2+y**2) / (2*std**2))

def G(x,y, std):
    return np.exp(-(x**2 + y**2)/(2*std**2))
```

# Edges are those places that the first derivative is not zero


note that the second derivative is zero both at the edges and also at those places that first derivative is zero.

So to filter out the constant regions, we set those places of laplaci that the first derivative is higher than a threshold

```python
filtered_ = LoG_filtered.copy()
filtered_[(np.abs(first_d)<60)]=1000
```

```python
plt.imshow(filtered_, cmap = 'gray')
plt.colorbar()
```

```python
fig, ax = plt.subplots(1)
ax.imshow(filtered_/5, cmap = 'prism')
ax.imshow(image, alpha=0.5, cmap='gray')
```

```python
def LoG_run(image, std):
    size=20
    kernel_x = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    kernel_y = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    first_d_x = signal.convolve(image, kernel_x, mode='same')
    first_d_y = signal.convolve(image, kernel_y, mode='same')
    first_d = (first_d_y**2 + first_d_x**2)**0.5
    x = np.arange(-size,size+1)
    mesh = np.meshgrid(x,x)
    LoG_kernel = LoG(mesh[0], mesh[1], std=std)
    LoG_filtered = signal.convolve(image, LoG_kernel, mode='same')
    #LoG_filtered = np.abs(LoG_filtered)
    LoG_filtered[np.abs(LoG_filtered)<200]=0
    LoG_filtered[np.abs(LoG_filtered)>50]=1000

    filtered_ = LoG_filtered.copy()
    filtered_[(np.abs(first_d)<150)]=1000   # You can vary limit of the firs_d

    fig, ax = plt.subplots(1)
    ax.imshow(filtered_/5, cmap = 'tab20c')
    ax.imshow(image, alpha=0.1, cmap='gray')
```

```python
LoG_run(image,3)
```
