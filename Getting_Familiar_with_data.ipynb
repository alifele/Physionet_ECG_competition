# Functions

```
def Get_Info(file_name, file_path='Training_WFDB/'):
  with open(file_path + file_name) as f:
    lines = f.readlines()

  age = lines[Age]
  sex = lines[Sex]
  d = lines[D]

  return age, sex, d
```

```
Get_Info('A6349.hea')
```

```
def data_loader(file_name, file_path = 'Training_WFDB/'):
  data = loadmat('Training_WFDB/A6349.mat')['val']
  return data
```
